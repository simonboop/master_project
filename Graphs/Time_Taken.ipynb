```r
#building an overall graph of time taken comparing bewtween experiments
#add coloum to each data frame so experiement can be identified
#add label to dataframes:
Change_Algorithm["Test"] = "Change_Algorithm"
Change_1prob["Test"] = "Change_1prob"
Changesd1["Test"] = "Changesd1"
Changesd2["Test"] = "Changesd2"
change1coverage["Test"] = "change1coverage"
Change_Coverage["Test"] = "Change_Coverage"
small_reads["Test"] = "small_reads"
Small_read_unrejectable["Test"] = "Small_read_unrejectable"
Match_time["Test"] = "Match_time"
Rej_time["Test"] = "Rej_time"
Reload_time["Test"] = "Reload_time"
Seq_Time["Test"] = "Seq_Time"
Change_number_amplicons["Test"] = "Change_number_amplicons"
#merge than and the work out means for each test and each algorithm
total = rbind(Change_Algorithm,Change_1prob,Changesd1,Changesd2,change1coverage,Change_Coverage,small_reads,Small_read_unrejectable,Match_time,Rej_time,Reload_time,Seq_Time)

tested = c("Change_Algorithm","Change_1prob","Changesd1","Changesd2","change1coverage","Change_Coverage","small_reads","Small_read_unrejectable","Match_time","Rej_time","Reload_time","Seq_Time")
algors = c("NoReadUntil","Prob_Model","Threshold","Min_Proport","Std_Restrict")

avgtotal= data.frame( AvgTime = numeric(0), Algorithm = character(0), Test =character(0),stringsAsFactors=FALSE)

#make mean table 
for (x in algors){
    for (y in tested){
        if (is.nan(with(total,mean(Time[Algorithm == x & Test == y]))) == FALSE){
            mean = as.integer(with(total,mean(Time[Algorithm == x & Test == y])))
            avgtotal[nrow(avgtotal) + 1, ] <- c(mean,x,y)
            }
    }
} 

avgtotal = transform(avgtotal, AvgTime = as.numeric(AvgTime))

#plot overall graph of time taken comparing bewtween experiments
ggplot(avgtotal,aes(Test,AvgTime, fill=Algorithm)) + 
geom_bar(stat= "identity", position= "dodge") +   
theme(axis.text.x=element_text(angle=90))+
ylab("Average Time Taken (s)") + xlab("Test Names") + ggtitle("Average Time Taken by Algorithm In Read Until Simulator \n To Reach Desired Coverage in Each Test")

```
